```python

import numpy
import os
import matplotlib.pyplot as plt

```

```python
### Read the data from the files from the folder

for file in os.listdir("."):
    if file.endswith(".npy"):
        data = numpy.load(file)
        print(data)
        plt.plot(data)
        if not os.path.exists(file + ".png"):
            plt.savefig(file + ".png")
        plt.show()
        ### if png not exist, save the plot as png
        
        plt.close()
```
